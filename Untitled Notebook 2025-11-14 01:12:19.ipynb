```python
import requests
import os
import tempfile
import shutil
from pyspark.sql.functions import col, mean, stddev, abs, monotonically_increasing_id

df = (
    spark.read.format("delta")
    .option("header", True)
    .option("inferSchema", True)
    .load("dbfs:/user/hive/warehouse/sample_anomaly_data")
)

numeric_col = "value"

dfIndexed = df.withColumn("index", monotonically_increasing_id()+1)

stats = dfIndexed.select(
    mean(col(numeric_col)).alias("mean"),
    stddev(col(numeric_col)).alias("stddev")
).collect()[0]

mean_val = stats["mean"]
stddev_val = stats["stddev"]

# Add z-score column and filter outliers
outliers_df = dfIndexed.withColumn(
    "z_score",
    (col(numeric_col) - mean_val) / stddev_val
).filter(abs(col("z_score")) > 2)

display(outliers_df)

spark.sql("""
    CREATE TABLE IF NOT EXISTS sample_anomaly_data_anomalies
    USING DELTA
    LOCATION 'dbfs:/user/hive/warehouse/sample_anomaly_data_anomalies'
""")

```
